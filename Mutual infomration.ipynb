```python
from jse import JoinSizeEstimation
from jse import UBS
from sklearn.metrics.cluster import mutual_info_score,adjusted_mutual_info_score
import pandas as pd

# from data import *

# population = pd.read_csv("Dataset/population.csv",sep='\t')
country = pd.read_csv("Dataset/country.csv")
# expectancy = pd.read_csv("Dataset/expectancy.csv",sep='\t')
# gnp = pd.read_csv("Dataset/gnp.csv",sep='\t')


population = country[['Code', 'Population']]
expectancy = country[['Code', 'LifeExpectancy']]
gnp = country[['Code', 'GNP']]
surfaceArea = country[['Code', 'SurfaceArea']]

expectancy.head()
```

```python
# def UBS(p: float, q: float, T: pd.DataFrame, J: str) -> pd.DataFrame

s2 = UBS(1,1,country,'Code')
m = adjusted_mutual_info_score(country['IndepYear'],country['IndepYear'])
print(m)
# ds = { 'country':country, 'expectancy': expectancy, 'gnp':gnp, 'population':population}
ds = {  'gnp':gnp, 'population':population,'surfaceArea':surfaceArea}
s_join = pd.merge(  expectancy,gnp, left_on='Code', right_on='Code')
mi_score = mutual_info_score(s_join['LifeExpectancy'],s_join.iloc[:, 2])
mi_score
```

```python
def mi_selection(main_df,target_attribute,join_key, data_set, max_sss, mi_collections,s_rates):
    mi_current = {}
    mi_diff = {}
    s_rates = {}
    
    if (max_sss >= 1):
        return 0;    
    
    for df_name, data_frame in ds.items():
        s1 = UBS(max_sss,1,main_df,join_key)
        s2 = UBS(max_sss,1,data_frame,join_key)
        s_join = pd.merge(
            s1, s2, left_on=join_key, right_on=join_key)
        mi_score = mutual_info_score(s_join[target_attribute],s_join.iloc[:, 2])
        mi_current.update({df_name:mi_score})
    
    max_score = 0
    max_name = ""   
    for mi_name, mi_score in mi_current.items():
        if mi_score >= max_score:
            max_score = mi_score
            max_name = mi_name
    print(max_score, max_name,max_sss)

    if(len(mi_collections)!= 0):
        for mi_name, mi_score in mi_collections.items():
            diff = mi_current.get(mi_name) - mi_score
            mi_diff.update({mi_name: diff})
        sorted_diff = dict(sorted(mi_diff.items(), key=lambda item: item[1]))
        print(sorted_diff)
        
    
    max_sss += 0.01
    mi_selection(main_df,target_attribute,join_key, data_set, max_sss, mi_current,s_rates)
    
    
    
mi_coll ={}
mi_selection(expectancy,'LifeExpectancy','Code', ds,0.01, mi_coll,{})   
        
```
